```python
import edward as ed
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

from edward.models import Normal, Categorical, Dirichlet, Empirical
```

```python
d = np.random.choice(np.arange(1,10))
n = np.random.choice(np.arange(100,10000))
```

```python
d, n
```

```python
X = np.empty((n,d))
```

```python
for i in range(d):
    if np.random.rand() > .5:
        k = np.random.choice(range(2,d))
        probs = np.random.dirichlet(np.ones(k))
        X[:,i] = np.random.choice(k, size=n, p = probs)
    else:
        mu_i = np.random.rand()*100
        sd_i = np.random.rand()*100
        X[:,i] = np.random.randn(n)*sd_i + mu_i
```

```python
X
```

```python
sns.set()
scattermatrix = sns.pairplot(pd.DataFrame(X))
```
