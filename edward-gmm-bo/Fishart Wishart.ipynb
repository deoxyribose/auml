```python
import edward as ed
import tensorflow as tf

from edward.models import Normal, WishartFull
import matplotlib.pylab as plt
%matplotlib inline
import numpy as np
```

```python
# MODEL
# given an n x d matrix X of n draws from a d-dimensional gaussian with covmat V, 
# the scatter matrix np.dot(X.T,X) ~ Wishart(n, V)
df = 10# must be a scalar greater than d
scale = np.array([[1, 0.2],[0.2, 3]]) # must be symmetric and positive definite
d = scale.shape[0]
z = WishartFull(df = df, scale = scale)
```

```python
from tensorflow.contrib.distributions.python.ops import operator_pd_cholesky, operator_pd_full
```

```python
#tmp = operator_pd_cholesky.OperatorPDCholesky(tf.nn.softplus(tf.random_normal([d,d]))).to_dense()
tmp = operator_pd_full.OperatorPDFull(tf.nn.softplus(tf.random_normal([d,d]))).to_dense()
```

```python
sess = tf.Session()
with sess.as_default():
    tmp.eval()
```

```python
# INFERENCE
#qz_M = tf.Variable(tf.random_normal([d,d]))
#qz_D = tf.diag(tf.nn.softplus(tf.Variable(tf.random_normal([d]))))
#qz_P = tf.cast(tf.matmul(qz_M,tf.matmul(qz_D,qz_M,transpose_b=True)), tf.float64)
qz_P = tf.cast(operator_pd_cholesky.OperatorPDCholesky(tf.nn.softplus(tf.random_normal([d,d]))).to_dense(), tf.float64)
qz_df = tf.cast(tf.nn.softplus(tf.Variable(tf.random_normal([]))+d), tf.float64)
qz = WishartFull(df=qz_df, scale=qz_P)

inference = ed.KLqp({z: qz})
#inference.initialize(n_iter=500, n_samples=20, n_minibatch=10, logdir='train')
inference.initialize(n_iter=500, n_samples=20)

tf.global_variables_initializer().run()
```

```python
for _ in range(inference.n_iter):
    info_dict = inference.update()
    #if _ % 100 == 0:
    #    tmp = qz.sample(sample_shape=(1000))
    #    plt.hist(tmp.eval(),50);
    #    plt.figure()
inference.print_progress(info_dict)
```
